# Log Query Notebook

```python
import pandas as pd

LOG_PATH = "../typhoon/output_data/master_summary_log.csv"
```

```python
""" Preview Dataframe """

log_df = pd.read_csv(LOG_PATH)
log_df
```

```python
""" Data that needs MANUAL check """

log_manual_ck_df = log_df[log_df['needs_manual_check'] == 1]
log_manual_ck_df
```

# 1. ดูรายการที่มีปัญหาเรื่อง "ผลรวมจำนวนบัตร" (Math Total Used)
> [จำนวนบัตรที่ใช้] = [บัตรดี] + [บัตรเสีย] + [ไม่เลือกใคร]

```python
math_used_fail = log_df[log_df['flag_math_total_used'] == True]
print(f"มีปัญหาผลรวมบัตร: {len(math_used_fail)} รายการ")
display(math_used_fail)
```

# 2. ดูรายการที่มีปัญหาเรื่อง "คะแนนรวมผู้สมัครไม่ตรงกับจำนวนบัตรดี" (Math Valid Score)

> [จำนวนบัตรดี] = [ผลรวมของคะแนนทุกพรรค/ผู้สมัครในตารางรวมกัน]

```python
math_score_fail = log_df[log_df['flag_math_valid_score'] == True]
print(f"มีปัญหาคะแนนรวมไม่ตรงกับบัตรดี: {len(math_score_fail)} รายการ")
display(math_score_fail)
```

# 3. ดูรายการที่ "ชื่อผู้สมัคร/พรรค อ่านเพี้ยน" (Name Mismatch)

```python
name_fail = log_df[log_df['flag_name_mismatch'] == True]
print(f"ชื่อไม่ตรงกับ Master List (Uthai Dist 2): {len(name_fail)} รายการ")
display(name_fail)
```

# 4. ดูรายการที่ "ผิดมากกว่า 1 อย่าง" (Multiple Failures)

```python
multi_fail = log_df[
    (log_df['flag_math_total_used'] == True) & 
    (log_df['flag_math_valid_score'] == True)
]
print(f"Emer ! ผิดทั้งบัตรและคะแนน: {len(multi_fail)} รายการ")
display(multi_fail)
```
