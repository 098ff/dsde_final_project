```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchWindowException, WebDriverException
import time
import urllib.parse
```

```python
# chrome_options = Options()
# chrome_options.add_argument("--headless")
driver = webdriver.Chrome()
base_url = "https://www.vote62.com/69/candidates/%E0%B8%AA%E0%B8%AA.%E0%B8%9A%E0%B8%B1%E0%B8%8D%E0%B8%8A%E0%B8%B5%E0%B8%A3%E0%B8%B2%E0%B8%A2%E0%B8%8A%E0%B8%B7%E0%B9%88%E0%B8%AD/" 
driver.get(base_url)
```

```python
links = driver.find_elements(By.CSS_SELECTOR, 'a[href*="/69/party/"]')
```

```python
party_list = []

for link in links:
    full_url = link.get_attribute('href')
    
    if full_url:
        party_name_raw = full_url.split('/')[-1]
        
        party_name = urllib.parse.unquote(party_name_raw)
        
        party_list.append({
            "name": party_name,
            "url": full_url
        })
```

```python
party_list
```

```python
len(party_list)
```

```python
results_data = []

for party in party_list:
    driver.get(party['url'])
    time.sleep(2)
    try :
        fb_element = driver.find_element(By.CSS_SELECTOR, 'a[href*="facebook.com"]')
        fb_url = fb_element.get_attribute('href')
        fb_id = fb_url.rstrip('/').split('/')[-1]
    except Exception:
        fb_url = None
        fb_id = None
    
    results_data.append({
        "name": party['name'],
        "facebook_id": fb_id
    })
```

```python
results_data
```

```python
cleaned_data = []

for item in results_data:
    fb_id = item['facebook_id']
    
    if fb_id == "vote62th":
        continue
    
    decoded_id = urllib.parse.unquote(fb_id)
    
    if "profile.php?id=" in decoded_id:
        clean_id = decoded_id.split("id=")[-1]
    elif "-" in decoded_id and any(char.isdigit() for char in decoded_id):
        clean_id = decoded_id.split("-")[-1]
    else:
        clean_id = decoded_id

    cleaned_data.append({
        'name': item['name'],
        'facebook_id': clean_id
    })
    
```

```python
cleaned_data
```

```python
len(cleaned_data)
```

```python
follower_results = []

for fb in cleaned_data :
  params = {
    "engine": "facebook_profile",
    "profile_id": fb['facebook_id'],
    "api_key": SERPAPI_API_KEY
  }

  search = serpapi.search(params)
  results = search.as_dict()
  followers = results.get("profile_results", {}).get("followers")

  follower_results.append({
    'name': fb['name'],
    'followers': followers
  })
```

```python
follower_results
```

```python
def transform_shorthand(value):
    if not value or not isinstance(value, str):
        return 0
    
    value = value.strip().upper() 
    
    multipliers = {
        'K': 1000,
        'M': 1000000,
        'B': 1000000000
    }
    
    unit = value[-1]
    if unit in multipliers:
        number = float(value[:-1].replace(',', ''))
        return int(number * multipliers[unit])
    
    try:
        return int(float(value.replace(',', '')))
    except:
        return 0
```

```python
for item in follower_results :
    num = item['followers']
    new_num = transform_shorthand(num)
    item['followers'] = new_num
```
