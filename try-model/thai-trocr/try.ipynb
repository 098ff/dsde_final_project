```python
import os
from PIL import Image
from pdf2image import convert_from_path

# ==========================================
# 1. ตั้งค่าไฟล์ PDF และหน้าที่จะเทส
# ==========================================
pdf_file_path = "อำเภอบ้านไร่/ตำบลแก่นมะกรูด/หน่วยเลือกตั้งที่ 1/ส.ส.5ทับ18 (บช.).pdf"  # เปลี่ยนชื่อไฟล์ให้ตรงกับของคุณ
page_number = 0  # 0 คือหน้าแรก (Index เริ่มที่ 0)

# ==========================================
# 2. แปลง PDF เป็น Image ก่อน
# ==========================================
try:
    print(f"กำลังแปลงไฟล์ {pdf_file_path} เป็นรูปภาพ (300 DPI)...")
    pages = convert_from_path(pdf_file_path, dpi=300)
    img = pages[page_number]
    
    print(f"✅ แปลงไฟล์สำเร็จ! ขนาดภาพ: กว้าง {img.width} x สูง {img.height} พิกเซล")

    # ==========================================
    # 3. ใส่พิกัดที่ต้องการเทส Crop
    # ==========================================
    # (ซ้าย, บน, ขวา, ล่าง) หรือ (x1, y1, x2, y2)
    # ** ให้คุณแก้เลขตรงนี้เพื่อหาพิกัดที่ตรงกับช่องที่ต้องการ **
    box = (1560, 1100, 1740, 1180)
    
    print(f"กำลังตัดภาพตามพิกัด: {box}")
    cropped_img = img.crop(box)

    # ==========================================
    # 4. เซฟและแสดงผลรูปที่ตัดแล้ว
    # ==========================================
    output_filename = 'test_crop_from_pdf.jpg'
    cropped_img.save(output_filename)
    
    print(f"✅ เซฟรูปที่ตัดแล้วเป็นไฟล์: {output_filename}")
    
    # คำสั่งนี้จะแสดงรูปภาพขึ้นมาให้เห็นใน Notebook เลย (สะดวกมาก!)
    display(cropped_img)

except FileNotFoundError:
    print(f"❌ ไม่พบไฟล์ '{pdf_file_path}' โปรดตรวจสอบว่าไฟล์อยู่ในโฟลเดอร์เดียวกับ Notebook หรือใส่ Path ถูกต้องไหม")
except Exception as e:
    print(f"❌ เกิดข้อผิดพลาด: {e}")
```
