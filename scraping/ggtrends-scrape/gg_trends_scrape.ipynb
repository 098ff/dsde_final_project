```python
party_names = [
    'ไทยทรัพย์ทวี', 'เพื่อชาติไทย', 'ใหม่', 'มิติใหม่', 'รวมใจไทย', 
    'รวมไทยสร้างชาติ', 'พลวัต', 'ประชาธิปไตยใหม่', 'เพื่อไทย', 'ทางเลือกใหม่', 
    'เศรษฐกิจ', 'เสรีรวมไทย', 'รวมพลังประชาชน', 'ท้องที่ไทย', 'อนาคตไทย', 
    'พลังเพื่อไทย', 'ไทยชนะ', 'พลังสังคมใหม่', 'สังคมประชาธิปไตยไทย', 'ฟิวชัน', 
    'ไทรวมพลัง', 'ก้าวอิสระ', 'ปวงชนไทย', 'วิชชั่นใหม่', 'เพื่อชีวิตใหม่', 
    'คลองไทย', 'ประชาธิปัตย์', 'ไทยก้าวหน้า', 'ไทยภักดี', 'แรงงานสร้างชาติ', 
    'ประชากรไทย', 'ครูไทยเพื่อประชาชน', 'ประชาชาติ', 'สร้างอนาคตไทย', 'รักชาติ', 
    'ไทยพร้อม', 'ภูมิใจไทย', 'พลังธรรมใหม่', 'กรีน', 'ไทยธรรม', 
    'แผ่นดินธรรม', 'กล้าธรรม', 'พลังประชารัฐ', 'โอกาสใหม่', 'เป็นธรรม', 
    'ประชาชน', 'ประชาไทย', 'ไทยสร้างไทย', 'ไทยก้าวใหม่', 'ประชาอาสาชาติ', 
    'พร้อม', 'เครือข่ายชาวนาแห่งประเทศไทย', 'ไทยพิทักษ์ธรรม', 'ความหวังใหม่', 'ไทยรวมไทย', 
    'เพื่อบ้านเมือง', 'พลังไทยรักชาติ', 'รวมพลัง', 'รักษ์ธรรม', 'สัมมาธิปไตย'
]
```

```python
import pandas as pd
from pytrends.request import TrendReq
import time
import random

pytrends = TrendReq(hl='th-TH', tz=360)


all_data = []

def chunker(seq, size):
    return (seq[pos:pos + size] for pos in range(0, len(seq), size))

print("กำลังดึงข้อมูลจาก Google Trends...")

for group in chunker(party_names, 5):
    try:
        pytrends.build_payload(group, cat=396, timeframe='2026-01-01 2026-03-31', geo='TH', gprop='')
        
        data = pytrends.interest_over_time()
        
        if not data.empty:
            data = data.drop(labels=['isPartial'], axis=1) 
            all_data.append(data)
            
        print(f"✅ ดึงข้อมูลกลุ่ม {group} สำเร็จ")
        wait_time = random.randint(60, 90)
        time.sleep(wait_time) 
        
    except Exception as e:
        print(f"❌ เกิดข้อผิดพลาดกับกลุ่ม {group}: {e}")

final_df = pd.concat(all_data, axis=1)

# Schema: [Date, Party_Name, Search_Interest]
melted_df = final_df.reset_index().melt(id_vars='date', var_name='party_name', value_name='search_interest')

# Export เป็น CSV
melted_df.to_csv('thailand_party_trends_3months.csv', index=False, encoding='utf-8-sig')
print("--- Export ไฟล์สำเร็จ: thailand_party_trends_3months.csv ---")
```

```python
import os

pytrends = TrendReq(hl='th-TH', tz=360)
file_name = 'thailand_party_trends_3months.csv'

already_scraped = []
if os.path.exists(file_name):
    existing_df = pd.read_csv(file_name)
    already_scraped = existing_df['party_name'].unique().tolist()
    print(f"พบข้อมูลเดิมในไฟล์: {len(already_scraped)} พรรค")

remaining_parties = [p for p in party_names if p not in already_scraped]

if not remaining_parties:
    print("✅ ข้อมูลครบถ้วนแล้ว ไม่ต้องดึงเพิ่ม")
else:
    print(f"เหลือที่ต้องดึงอีก {len(remaining_parties)} พรรค...")

    all_data = []

    def chunker(seq, size):
        return (seq[pos:pos + size] for pos in range(0, len(seq), size))

    for group in chunker(remaining_parties, 3):
        try:
            pytrends.build_payload(group, cat=396, timeframe='2026-01-01 2026-03-31', geo='TH', gprop='')
            data = pytrends.interest_over_time()
            
            if not data.empty:
                data = data.drop(labels=['isPartial'], axis=1)
                all_data.append(data)
                
            print(f"✅ ดึงข้อมูลกลุ่ม {group} สำเร็จ")
            time.sleep(random.randint(120, 180)) 
            
        except Exception as e:
            print(f"❌ เกิดข้อผิดพลาดกับกลุ่ม {group}: {e}")

    if all_data:
        new_df = pd.concat(all_data, axis=1)
        new_melted = new_df.reset_index().melt(id_vars='date', var_name='party_name', value_name='search_interest')
        
        if os.path.exists(file_name):
            final_df = pd.concat([existing_df, new_melted], ignore_index=True)
        else:
            final_df = new_melted
            
        final_df.to_csv(file_name, index=False, encoding='utf-8-sig')
        print(f"--- อัปเดตไฟล์สำเร็จ! ตอนนี้มีข้อมูลรวม {final_df['party_name'].nunique()} พรรค ---")
    else:
        print("ไม่มีข้อมูลใหม่ที่ถูกดึงมา")
```
