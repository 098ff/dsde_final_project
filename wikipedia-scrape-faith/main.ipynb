```python
import requests
import pandas as pd
import numpy as np 
from bs4 import BeautifulSoup

# --- 1. ฟังก์ชันพิเศษสำหรับอ่านตาราง Wikipedia แบบทะลุทะลวง Rowspan ---
def parse_html_table(table):
    rows = table.find_all('tr')
    
    # หาจำนวนคอลัมน์ที่แท้จริง
    max_cols = 0
    for tr in rows:
        max_cols = max(max_cols, sum(int(c.get('colspan', 1)) for c in tr.find_all(['th', 'td'])))
        
    grid = [[None] * max_cols for _ in range(len(rows))]
    
    # เติมข้อมูลลง Grid โดยคำนึงถึงเซลล์ที่โดน Merge (Rowspan/Colspan)
    for r_idx, tr in enumerate(rows):
        c_idx = 0
        for cell in tr.find_all(['th', 'td']):
            while c_idx < max_cols and grid[r_idx][c_idx] is not None:
                c_idx += 1
            if c_idx >= max_cols:
                break
            
            text = cell.get_text(strip=True)
            rowspan = int(cell.get('rowspan', 1))
            colspan = int(cell.get('colspan', 1))
            
            # ถมข้อมูลลงในช่องว่างทุกช่องที่เซลล์นี้ครอบคลุม
            for r in range(r_idx, min(r_idx + rowspan, len(rows))):
                for c in range(c_idx, min(c_idx + colspan, max_cols)):
                    grid[r][c] = text
                    
            c_idx += colspan
    return grid
# -----------------------------------------------------------------

set_number = 23
print(f"--- Parliament {set_number} (Ultimate Fix for Multi-Member) ---")
url = URLs[set_number]
response = requests.get(url, headers=headers)
data_23 = []

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    wiki_tables = soup.find_all('table', {'class': 'wikitable'})
    print(f"Found {len(wiki_tables)} tables.")

    for i, table in enumerate(wiki_tables):
        try:
            all_trs = table.find_all('tr')
            if not all_trs: continue

            # เช็คว่าเป็นตาราง สส. หรือไม่
            is_target_table = False
            for tr in all_trs[:3]:
                header_text = "".join([cell.get_text(strip=True) for cell in tr.find_all(['th', 'td'])])
                if 'เขต' in header_text and 'พรรค' in header_text:
                    is_target_table = True
                    break
            
            if not is_target_table: continue
            
            # เรียกใช้ฟังก์ชันพิเศษของเรา
            grid = parse_html_table(table)
            
            # ตอนนี้ข้อมูลใน grid จะถูกเติมเต็มทุกช่อง (ไม่มีเบี้ยวซ้าย/ขวาแล้ว)
            df = pd.DataFrame(grid)
            
            if len(df.columns) >= 5: 
                # เนื่องจาก grid เรียงข้อมูลถูกต้อง 100% แล้ว
                # เราสามารถดึง Col 0, 1, 2, 4 ได้เลย ไม่ว่าตารางจะมีกี่แถบสีก็ตาม
                df_cleaned = df[[0, 1, 2, 4]].copy()
                df_cleaned.columns = ['province', 'district', 'name', 'party']

                # กรองเอาเฉพาะแถวที่เป็นตัวเลขเขต
                mask = df_cleaned['district'].str.isnumeric().fillna(False)
                df_cleaned = df_cleaned[mask]
                
                data_23.append(df_cleaned)
                print(f"Table {i} success! Cleaned {len(df_cleaned)} MPs.")

        except Exception as e:
            print(f"Skip Table {i} due to: {e}")

    if data_23:
        final_df_23 = pd.concat(data_23, ignore_index=True)
        print("\n--- Final Data Preview ---")
        print(final_df_23.head(15)) 
        print(f"\nTotal MPs Extracted: {len(final_df_23)}")
        
        if (len(final_df_23) == target_district[set_number]):
            print(f"Parliament {set_number} MATCHED. ✅")
        else:
            print(f"Parliament {set_number} NOT MATCHED. ❌")
            
        filename = f"parliament_{set_number}_district.csv"
        final_df_23.to_csv(filename, index=False, encoding='utf-8-sig')
        print(f"Exported to {filename}")
else:
    print(f"Request Error. Status Code: {response.status_code}")
```

# Partylist

```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup

sets_to_scrape = [27, 26, 25, 24, 23, 22, 21] 

for set_number in sets_to_scrape:
    if set_number not in URLs:
        continue
        
    print(f"\n--- Parliament {set_number} (Party-List) ---")
    url = URLs[set_number]
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Request Error. Status Code: {response.status_code}")
        continue
        
    soup = BeautifulSoup(response.text, 'html.parser')
    wiki_tables = soup.find_all('table', {'class': 'wikitable'})
    
    data_partylist = []
    
    for i, table in enumerate(wiki_tables):
        try:
            name_col_idx = -1
            for tr in table.find_all('tr')[:3]:
                cells = tr.find_all(['th', 'td'])
                row_texts = [c.get_text(strip=True) for c in cells]
                
                for idx, text in enumerate(row_texts):
                    if 'รายชื่อ' in text or 'รายนาม' in text:
                        name_col_idx = idx
                        break
                if name_col_idx != -1:
                    break
            
            if name_col_idx == -1:
                continue 
                
            party_name = "Unknown Party"
            prev_header = table.find_previous(['h2', 'h3', 'h4', 'h5'])
            if prev_header:
                raw_text = prev_header.get_text(strip=True)
                clean_party = re.sub(r'\s*\(\d+\)\s*', '', raw_text)
                clean_party = re.sub(r'\[.*?\]', '', clean_party).strip()
                
                if 'บัญชีรายชื่อ' not in clean_party:
                    party_name = clean_party
                    
            for tr in table.find_all('tr'):
                cells = tr.find_all(['th', 'td'])
                row_data = [c.get_text(strip=True) for c in cells]
                
                if not row_data or len(row_data) <= name_col_idx:
                    continue
                    
                list_num = row_data[0] 
                
                if list_num.isnumeric():
                    name = row_data[name_col_idx]
                    
                    name = re.sub(r'\[.*?\]', '', name).strip()
                    
                    if name:
                        data_partylist.append({
                            'party': party_name,
                            'list_number': int(list_num),
                            'name': name
                        })
                        
        except Exception as e:
            print(f"Skip Table {i} due to: {e}")
            
    if data_partylist:
        df = pd.DataFrame(data_partylist)
        print(f"Extracted {len(df)} Party-list MPs.")
        print(df.head(3))
        print("...")
        print(df.tail(3)) 
        
        filename = f"parliament_{set_number}_partylist.csv"
        df.to_csv(filename, index=False, encoding='utf-8-sig')
        print(f"Exported -> {filename}")
    else:
        print("No Partylist MPs found.")
```

```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup

sets_to_scrape = [24, 23, 22, 21] 

for set_number in sets_to_scrape:
    if set_number not in URLs:
        continue
        
    print(f"\n========== Parliament {set_number} (Party-List / Proportional) ==========")
    url = URLs[set_number]
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Request Error: {response.status_code}")
        continue
        
    soup = BeautifulSoup(response.text, 'html.parser')
    wiki_tables = soup.find_all('table', {'class': 'wikitable'})
    
    data_partylist = []
    
    for i, table in enumerate(wiki_tables):
        try:
            party_name = "Unknown Party"
            for header in table.find_all_previous(['h2', 'h3', 'h4', 'h5']):
                raw_text = header.get_text(strip=True)
                # ตัดตัวเลขในวงเล็บ (61) ออก
                clean_party = re.sub(r'\s*\(\d+\)\s*', '', raw_text)
                clean_party = re.sub(r'\[.*?\]', '', clean_party).strip()
                
                if 'พรรค' in clean_party and 'รายนาม' not in clean_party and 'รายชื่อ' not in clean_party:
                    party_name = clean_party
                    break 
            
            is_mp_table = False 

            for tr in table.find_all('tr'):
                cells = tr.find_all(['th', 'td'])
                row_data = [c.get_text(strip=True) for c in cells]
                
                if not row_data:
                    continue
                    
                col_0 = row_data[0] 
                
                if col_0.isnumeric():
                    is_mp_table = True
                    
                    name = ""
                    for cell_text in row_data[1:]:
                        clean_text = re.sub(r'\[.*?\]', '', cell_text).strip()
                        if len(clean_text) > 2 and "ลาออก" not in clean_text:
                            name = clean_text
                            break 
                    
                    if name:
                        data_partylist.append({
                            'party': party_name,
                            'list_number': int(col_0),
                            'name': name
                        })
                        
            if not is_mp_table:
                 continue
                 
        except Exception as e:
            print(f"Skip Table {i} due to: {e}")
            
    if data_partylist:
        df = pd.DataFrame(data_partylist)
        print(f"✅ Extracted {len(df)} Party-list/Proportional MPs.")
        print(df.head(3))
        print("...")
        print(df.tail(3))
        
        filename = f"parliament_{set_number}_partylist.csv"
        df.to_csv(filename, index=False, encoding='utf-8-sig')
        print(f"Exported -> {filename}")
    else:
        print(f"❌ No Partylist MPs found for Parliament {set_number}.")
```
