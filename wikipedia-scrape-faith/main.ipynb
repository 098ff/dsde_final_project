```python
import requests
import pandas as pd
import numpy as np 
import io
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

set_numbers = [27, 26, 25, 24, 23, 22, 21]

wrong_district_set = []

for set_number in set_numbers:
    print(f"--- Parliament {set_number} ---")
    url = URLs[set_number]
    response = requests.get(url, headers=headers)
    data[set_number] = []

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        wiki_tables = soup.find_all('table', {'class': 'wikitable'})
        print(f"Found {len(wiki_tables)} tables.")

        for i, table in enumerate(wiki_tables):
            try:
                all_trs = table.find_all('tr')
                if not all_trs:
                    continue

                is_target_table = False
                for tr in all_trs[:3]:
                    header_text = "".join([cell.get_text(strip=True) for cell in tr.find_all(['th', 'td'])])
                    if 'เขต' in header_text and 'พรรค' in header_text:
                        is_target_table = True
                        break
                
                if not is_target_table:
                    continue

                rows = []
                target_max_cols = 7

                for tr in all_trs:
                    cells = tr.find_all(['th', 'td'])
                    row = [cell.get_text(strip=True) for cell in cells]
                    
                    if not row:
                        continue

                    if len(row) < target_max_cols:
                        diff = target_max_cols - len(row)
                        row = [np.nan] * diff + row
                    
                    rows.append(row)
                
                if rows:
                    df = pd.DataFrame(rows)
                    
                    # district should have 5 columns in header (7 columns in data) 
                    if len(df.columns) == target_max_cols:
                        # col 0 = Province
                        # col 1 = District
                        # col 2 = Name + Surname
                        # col 4 = Party
                        df_cleaned = df[[0, 1, 2, 4]].copy()
                        df_cleaned.columns = ['province', 'district', 'name', 'party']
                        
                        # forward fill province name from upper row
                        df_cleaned['province'] = df_cleaned['province'].replace('', np.nan).ffill()

                        # clean header out                    
                        mask = df_cleaned['district'].str.isnumeric().fillna(False)
                        df_cleaned = df_cleaned[mask]
                        
                        data[set_number].append(df_cleaned)
                        print(f"Table {i} success! Cleaned {len(df_cleaned)} MPs.")

            except Exception as e:
                print(f"Skip Table {i} due to: {e}")


        if data[set_number]:
            data[set_number] = pd.concat(data[set_number], ignore_index=True)
            print("\n--- Final Data Preview ---")
            print(data[set_number].head(10))
            print(f"Total MPs Extracted: {len(data[set_number])}")
            
            if (len(data[set_number]) == target_district[set_number]):
                print(f"Parliament {set_number} MATCHED. ✅")
                filename = f"parliament_{set_number}_district.csv"
                data[set_number].to_csv(filename, index=False, encoding='utf-8-sig')
                print(f"Exported to {filename}")
            else:
                print(f"Parliament {set_number} NOT MATCHED. ❌")
                wrong_district_set.append(set_number)

    else:
        print(f"Request Error. Status Code: {response.status_code}")

print("\n--- Summary ---")
print(f"Sets with missing/extra MPs: {wrong_district_set}")
```
